```python
import sys
import numpy as np
```

```python
variable = "tau_theta"
start_val = float(0)
end_val = float(1000)
step_val = float(10) 
```

```python
f = open('parameters_default.txt',"r")
lines = f.readlines()
params = []
for x in lines:
    params.append(x.split())
f.close()
params = np.array(params)
```

```python
params
```

```python
varIndex=90
for m in range (0,23):
    if(variable==params[m,0]):
        varIndex=m
        break
```

```python
params2 = params
values = np.arange(start_val, end_val+step_val , step_val)
```

```python
i=0
for n in values: 
    params2[varIndex,2]=n
    np.savetxt('parameter/'+ 'parameters_'+variable+'_'+str(i)+'.txt', params2, fmt='%s')
    i+=1

#calculating number of simulations required 
simNumber = (end_val - start_val)/step_val+1
print("Put simNumber as: "+str(simNumber));
```
